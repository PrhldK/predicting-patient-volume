```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
df = pd.read_csv("final_predictions.csv")
```

```python
df.index = df["date"]
```

```python
df.head()
```

```python
df.index = pd.to_datetime(df.index)
```

```python
fig, ax = plt.subplots(figsize=(15,6))
plt.scatter(df.index, df["y_actual"], marker='o', color='black', label='Actual')
plt.scatter(df.index, df["dy_baseline"], marker='3', color='blue', label='Dynamic Baseline')
plt.scatter(df.index, df["dy_ts"], marker='4', color='purple', label='Dynamic TS')
plt.scatter(df.index, df["dy_nn"], marker='1', color='green', label='Dynamic NN')
plt.scatter(df.index, df["dy_rnn"], marker='2', color='orange', label='Dynamic RNN')
plt.legend(loc=1)
plt.xlim((pd.datetime(2017,12,31), pd.datetime(2018,9,21)))
plt.xlabel("Date")
plt.ylabel("Sick Visits/FTP")
plt.title("Dynamic Model Predictions For All Visits")
plt.show()
```

```python
desc = pd.Series(df["dy_baseline_r"] - df["y_resp"]).describe()
iqr15 = 1.5 * (desc[6] - desc[4])
iqr_above = np.array([iqr15 for i in range(len(df))])
iqr_below = np.array([-1 * iqr15 for i in range(len(df))])

fig, ax = plt.subplots(figsize=(15,12))
plt.subplot(2, 1, 1)
plt.scatter(df.index, df["dy_baseline_r"] - df["y_resp"], marker='3', color='blue', label='Dynamic Baseline')
plt.scatter(df.index, df["dy_ts_r"] - df["y_resp"], marker='4', color='purple', label='Dynamic TS')
plt.scatter(df.index, df["osa_ts_r"] - df["y_resp"], marker='.', color='purple', label='OSA TS')
plt.scatter(df.index, df["dy_nn_r"] - df["y_resp"], marker='1', color='green', label='Dynamic NN')
plt.scatter(df.index, df["osa_nn1_r"] - df["y_resp"], marker='.', color='green', label='OSA NN-1')
plt.scatter(df.index, df["dy_rnn_r"] - df["y_resp"], marker='2', color='orange', label='Dynamic RNN')
plt.scatter(df.index, df["osa_rnn7_r"] - df["y_resp"], marker='.', color='orange', label='OSA RNN-7')
plt.plot(df.index, iqr_above, '-', color='teal', label='Â±1.5 * Baseline IQR range')
plt.plot(df.index, iqr_below, '-', color='teal')
plt.axvline(dt.datetime(2018,1,14), color='r', linestyle=':', label='Approximate Range of Outbreak')
plt.axvline(dt.datetime(2018,3,1), color='r', linestyle=':')
plt.legend(loc=0)
plt.xlim((pd.datetime(2017,12,31), pd.datetime(2018,9,21)))
plt.xlabel("Date")
plt.ylabel("Error (Sick Visits/FTP)")
plt.title("Dynamic and OSA Model Errors on Respiratory Visits for January-September 2018")

plt.subplot(2, 1, 2)
plt.scatter(df.index, df["dy_baseline_r"] - df["y_resp"], marker='3', color='blue', label='Dynamic Baseline')
plt.scatter(df.index, df["dy_ts_r"] - df["y_resp"], marker='4', color='purple', label='Dynamic TS')
plt.scatter(df.index, df["osa_ts_r"] - df["y_resp"], marker='.', color='purple', label='OSA TS')
plt.scatter(df.index, df["dy_nn_r"] - df["y_resp"], marker='1', color='green', label='Dynamic NN')
plt.scatter(df.index, df["osa_nn1_r"] - df["y_resp"], marker='.', color='green', label='OSA NN-1')
plt.scatter(df.index, df["dy_rnn_r"] - df["y_resp"], marker='2', color='orange', label='Dynamic RNN')
plt.scatter(df.index, df["osa_rnn7_r"] - df["y_resp"], marker='.', color='orange', label='OSA RNN-7')
plt.plot(df.index, iqr_above, '-', color='teal', label='Â±1.5 * Baseline IQR range')
plt.plot(df.index, iqr_below, '-', color='teal')
plt.legend(loc=0)
plt.xlim((pd.datetime(2018,1,14), pd.datetime(2018,3,1)))
plt.xlabel("Date")
plt.ylabel("Error (Sick Visits/FTP)")
plt.title("Dynamic and OSA Model Errors on Respiratory Visits for the Approximate Range of Outbreak")

plt.show()
```
